```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
''' Create a DataFrame from CSV File'''

df = pd.read_csv('fifa21_male2.csv', low_memory=False)

# Provide the number of rows and columns in this dataset

num_rows = df.shape[0] #Provide the number of rows in the dataset
num_cols = df.shape[1] #Provide the number of columns in the dataset

print('There are {} rows in the dataset!'.format(num_rows))
print('There are {} columns in the dataset!'.format(num_cols))
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

df.describe()
```

```python
df.columns
print(df.columns.tolist())
```

```python
# Which columns had no missing values?

no_nulls = set(df.columns[np.sum(df.isnull()) == 0]) #Provide a set of columns with 0 missing values.
print(no_nulls)
```

```python
#  Which columns have the most missing values?

most_25_missing_cols = set(df.columns[np.sum(df.isnull())/df.shape[0] > .25]) #Provide a set of columns with more than 25% of the values missing
most_50_missing_cols = set(df.columns[np.sum(df.isnull())/df.shape[0] > .50]) #Provide a set of columns with more than 50% of the values missing
most_75_missing_cols = set(df.columns[np.sum(df.isnull())/df.shape[0] > .75]) #Provide a set of columns with more than 75% of the values missing

print('The columns with more than 25% of the values: ', most_25_missing_cols)
print('The columns with more than 50% of the values: ', most_50_missing_cols)
print('The columns with more than 75% of the values: ', most_75_missing_cols)
```

```python
df = df.drop(['Loan Date End'], axis = 1)

df.columns
print(df.columns.tolist())
```

```python
''' Sort the Players by Overall Score '''

best_players = df.sort_values(['OVA'],ascending=[False]).head(100)

selected = best_players[['Name', 'OVA']]
selected.head(10)
```

```python
''' Sort the Players by Overall Score and No of Hits '''

best_players = df.sort_values(['OVA', 'Hits'],ascending=[False, False]).head(100)

selected = best_players[['Name', 'OVA', 'Hits']]
topdf = selected.head(3)
```

```python
top3 = [['L. Messi', 93, 830], ['Cristiano Ronaldo', 92, 750], ['Neymar Jr', 91, 546]]
top3df = pd.DataFrame(top3, columns=["Name","OVA","Hits"])
print(top3df)
print(topdf)
```

```python
''' Plot a Bar Chart for the Top 3 Plyers and Show their Overall Score and No of Hits'''

top3df.plot(x='Name', y=['OVA', 'Hits'], kind='bar');
```

```python
''' Preferred Foot '''

foot_vals = df['foot'].value_counts() #Provide a pandas series of the counts for each Preferred Foot
print(foot_vals)
```

```python
''' Which Countries do FIFA Players come from?  '''

nationality_vals = df['Nationality'].value_counts() #Provide a pandas series of the counts for each Nationality status
print('There are {} different Nationalities in FIFA 2021!'.format(nationality_vals.count()))
```

```python
# The below should be a bar chart of the 10 largest population in the Nationality category 
(nationality_vals.nlargest(10)).plot(kind="bar");
plt.title("Where FIFA players come from?");
# print(nationality_vals)
```
